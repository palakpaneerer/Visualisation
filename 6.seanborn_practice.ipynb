# 1 前準備

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plot_df = pd.read_csv("csv_data/sample_df.csv")
plot_df.head()
```

# 2 1次元のプロット

## 2.1 ヒストグラム

```python
plt.figure(figsize=(5,3))
plt.title('Japanese socre', fontsize=10)
plt.xlabel('Score', fontsize=10)
plt.ylabel('Counts', fontsize=10)
plt.tick_params(labelsize=10)
plt.ylim(0,10)

sns.histplot(x='Japanese', data=plot_df, hue='Sex', alpha=0.5)
plt.show()
```

```python
g = sns.FacetGrid(plot_df, col='Sex')
g.map_dataframe(sns.histplot, x='Japanese')
plt.show()
```

## 2.2 棒グラフ

```python
plt.figure(figsize=(5,3))
plt.title('input title', fontsize=10)

sns.countplot(x='Sex', data=plot_df)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title('Counts by sex')

sns.countplot(x='Sex', data=plot_df, hue='Class')
plt.show()
```

```python
g = sns.FacetGrid(plot_df, col='Class')
g.map_dataframe(sns.countplot, x='Sex')
plt.show()
```

# 3 2次元プロット

## 3.1 散布図

```python
plt.figure(figsize=(5,3))
plt.title('Scatter plot', fontsize=10)
sns.scatterplot(x='Japanese', y='Math', data=plot_df)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title('Strip plot', fontsize=10)
sns.stripplot(x='Sex', y='Math', data=plot_df)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
sns.scatterplot(x='Japanese', y='Math', data=plot_df, hue='Sex')
plt.legend(loc='upper right')
plt.show()
```

```python
g = sns.FacetGrid(plot_df, col='Class')
g.map_dataframe(sns.scatterplot, x='Japanese', y='Math')
plt.show()
```

## 3.2 折れ線グラフ

```python
plt.figure(figsize=(5,3))
plt.title('Math score', fontsize=10)
sns.lineplot(x='StudentID', y='Math', data=plot_df, hue='Sex')
plt.legend(loc='upper right')
plt.show()
```

```python
g = sns.FacetGrid(plot_df, col='Class')
g.map_dataframe(sns.lineplot, x='StudentID', y='Math')
plt.show()
```
