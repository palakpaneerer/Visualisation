```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

# 1 基本的な可視化

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True)
plt.tick_params(labelsize = 10)
plt.show()
```

# 2 1次元プロット

## 2.1 ヒストグラム (1次元)

```python
x_hist = np.random.normal(50, 10, 100)
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)
plt.xlim(0, 100)
plt.ylim(0, 50)
plt.grid(True)
plt.tick_params(labelsize = 10)

plt.hist(x_hist)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)
plt.xlim(0, 100)
plt.ylim(0, 50)
plt.grid(True)
plt.tick_params(labelsize = 10)

plt.hist(x_hist, bins=20, color="red", alpha=0.1, ec='black')
plt.show()
```

## 2.2 棒グラフ (1次元)

```python
x_bar = np.random.choice(["Male", "Female"], 100)
```

```python
count_unique = np.unique(x_bar, return_counts=True)
axis_x = range(len(count_unique))
```

```python
print(count_unique)
print()
print(count_unique[0])
print(count_unique[1])
print(axis_x)
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)

plt.tick_params(labelsize = 10)
plt.xticks(axis_x, count_unique[0])

plt.bar(axis_x, count_unique[1])
plt.show()
```

# 3 2次元プロット

## 3.1 散布図 (2次元)

```python
x_plot = np.random.normal(50, 10, 100)
y_plot = np.random.normal(x_plot, 10, 100)
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)

plt.tick_params(labelsize = 10)

plt.scatter(x_plot, y_plot)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.title("input title", fontsize=10)
plt.xlabel("input xlabel", fontsize=10)
plt.ylabel("input ylabel", fontsize=10)

plt.tick_params(labelsize = 10)

plt.scatter(x_plot, y_plot,
            color="blue", s=10, marker="x")
plt.show()
```

## 3.2 折れ線グラフ

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)
ax4 = fig.add_subplot(2, 3, 4)
ax5 = fig.add_subplot(2, 3, 5)

ax2.hist(x_hist, bins=20, color="red",
         alpha=0.1, ec='black')

ax4.set_title("sample_title")
ax4.set_ylabel("samle_ylabel")
ax4.set_xlabel("samle_xlabel")
ax4.set_xlim(0, 100)
ax4.set_ylim(0, 50)
ax4.hist(x_hist, bins=20, color="red",
         alpha=0.1, ec='black')

plt.tight_layout()
plt.show()
```

```python
h = 2
w = 3

fig = plt.figure(figsize=(10, 5))
for n in range(h*w):
    ax = fig.add_subplot(h, w, n+1)
    # ここにプロットの詳細
    ax.set_title("sample_title")
    ax.set_ylabel("samle_ylabel")
    ax.set_xlabel("samle_xlabel")
    
plt.tight_layout()
plt.show()
```
